```python
#FRONTEND
import geopandas as gpd
import ipywidgets as widgets
from IPython.display import display
import requests
import json
# new packages for the frontend
from dash import Dash, html, dcc
import plotly.express as px
```

USER ACCESS/REGISTRATION

```python
#necessary widgets
username = widgets.Text(
    placeholder='Username/Email',
    description='Username:',
    disabled=False)
email = widgets.Text(
    placeholder='Email',
    description='Email:')
password = widgets.Password(
    value='password',
    placeholder='Enter password',
    description='Password:',
    disable=False)
login_button = widgets.Button(description='Login', button_style='success')
signin_button = widgets.Button(description='Sign in', button_style='info')
output = widgets.Output()

# SIGN IN FUNCTION
def signin(b):
    new_user_data = {
        "username": username.value,
        "email": email.value,
        "password": password.value,
        } 
    try:
        print("Sending data:", new_user_data)
        
        user_signin = requests.post(
        url = "http://localhost:5000/api/signin",
        json = new_user_data
        ) 
        
        with output:
            output.clear_output()
            if user_signin.status_code == 200:
                print("Sign in successful.")
            elif user_signin.status_code == 409:
                print("User already exists.")
            elif user_signin.status_code == 400:
                print("All fields are required.")
            else:
                print(f"Error: {user_signin.status_code} - {user_signin.text}")
    except requests.exceptions.RequestException as e:
        with output:
            output.clear_output()
            print(f"Connection Error: {e}")


def display_signin_form():
    output.clear_output()
    with output:
        display(username, email, password, signin_button)
    signin_button._click_handlers.callbacks.clear()
    signin_button.on_click(signin)


# LOGIN FUNCTION
def login(b):
    user_data = {
        "username": username.value,
        "password": password.value,
        }
    try:
        user_login = requests.post(
        url = "http://localhost:5000/api/login",
        json = user_data
        ) 
        with output:
            output.clear_output()
            if user_login.status_code == 200:
                print("Login successful.")
            elif user_login.status_code == 403:
                print("Wrong password.")
            elif user_login.status_code == 401:
                print("Non registered user. Please sign in.")
                display_signin_form()
            else:
                print(f"Error: {user_login.status_code} - {user_login.text}")
    except requests.exceptions.RequestException as e:
        with output:
            output.clear_output()
            print(f"Connection Error: {e}")  



login_button.on_click(login)
display(username, password, login_button, output)
```
